```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
from collections import defaultdict
```

```python
g = defaultdict(list)
gr = nx.Graph()


# gr.add_edge(1,2)
# edge = ("d", "e")
# gr.add_edge(*edge)
# edge = ("a", "b")
# gr.add_edge(*edge)

# print("Nodes of graph: ")
# print(G.nodes())
# print("Edges of graph: ")
# print(G.edges())
```

```python
it = 0
for pers in data:
    it += 1
#     if it > 4:
#         break
#     print(pers['affilated'])
    for link in pers['affilated']:
        start = pers['id']
        finish = link['person']['id']
        gr.add_node(start)
        gr.add_node(finish)
        g[start].append(finish)
        g[finish].append(start)
        gr.add_edge(start, finish)
        
```

```python
for key in g.keys():
    print('key =', key, ' len =', len(g[key]))
```

```python
plt.figure(figsize=(21,21)) 

nx.draw(gr, node_size=1)
plt.savefig("simple_path.png") # save as png
plt.show() # display
```

```python
colored = set()
def analyze_connection(node):
    for newnode in g[node]:
        if newnode not in colored:
            colored.add(node)
            analyze_connection(newnode)
```
